```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
%matplotlib inline
x=np.array([5,6,7])
y=np.array([1,2,4])
plt.plot(x,y,label="linear line",color='g')
plt.title("sample linear graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
%matplotlib inline
x=np.array([1,2,3])
y=np.array([2,3,5])
plt.scatter(x,y,label="Scatter plot",color='m',marker="X")
plt.title("Simple Scatter Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True,color='k')
plt.legend()
plt.show()
```
